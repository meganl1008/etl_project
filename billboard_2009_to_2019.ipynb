````python
date_list = [data for data in bill_board_2009_to_2019['date'].values]

year_list = []
for date in date_list:
    year = date.split("-")[0]
    year_list.append(year)
````

````python
bill_board_2009_to_2019
````

````python
bill_board_2009_to_2019.insert(3, "year", year_list, True)
bill_board_2009_to_2019['year'].value_counts()
````

````python
bill_board_2009_to_2019['year'] = bill_board_2009_to_2019['year'].astype(int)
billboard_2010_to_2019 = bill_board_2009_to_2019.loc[(bill_board_2009_to_2019['year'] > 2009)]

billboard_2010_to_2019['year'].value_counts()
````

````python
year_list = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
df_year_list = []

for year in year_list:
    year_df = billboard_2010_to_2019.loc[billboard_2010_to_2019['year'] == year]    
    artist_by_year = year_df[['artist_name', 'year']]
    df_year_list.append(artist_by_year)
    
new_df_list = []
for df in df_year_list:
    table = pd.pivot_table(df, index=['artist_name'], aggfunc = 'count')
    new_df = table.reset_index()
    new_df = new_df.sort_values(by='year', ascending=False)
    new_df = new_df.rename(columns={"year": "count"})
    new_df_list.append(new_df)

top_artist_list = []
max_count_list = []

for i in new_df_list:
    max_count = i['count'].max()
    top_artists = i.loc[i['count'] == max_count]
    top_artists = top_artists[['artist_name']]
    top_artist_list.append(top_artists)
    
full_df_list = []
for i in range(len(top_artist_list)):
    top_artist_list[i]['year'] = year_list[i]
    full_df_list.append(top_artist_list)
````

````python
#artist that showed up most in weekly top 100 list for each year 
concat_df_list = [full_df_list[i][0], full_df_list[i][1], full_df_list[i][2], full_df_list[i][3], 
                  full_df_list[i][4], full_df_list[i][5], full_df_list[i][6], full_df_list[i][7], 
                  full_df_list[i][8], full_df_list[i][9]]

top_artists = pd.concat(concat_df_list)
top_artists = top_artists.reset_index().drop(columns = 'index')
top_artists
````
